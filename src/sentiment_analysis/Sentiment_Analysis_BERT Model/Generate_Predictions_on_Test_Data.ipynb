```python
# Import necessary libraries
from torch import nn
from transformers import BertTokenizer, BertForSequenceClassification, BertModel

import pyspark.sql.types as pyspark_types
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, FloatType, ArrayType, MapType
from pyspark.sql.functions import udf, collect_list, explode, col, rank, concat_ws, row_number



# Torch ML libraries
import transformers
from transformers import BertModel, BertTokenizer, AdamW, get_linear_schedule_with_warmup
import torch
from torch import nn, optim
from torch.utils.data import Dataset, DataLoader

# Misc.
import warnings
warnings.filterwarnings('ignore')


# Load Test Data
dataset_fp = "/content/drive/MyDrive/CS5344_Project/Dataset/Electronics_preprocessed.json"

# Define the schema for the Spark DataFrame
json_schema_data = pyspark_types.StructType([
    StructField("overall", FloatType(), True),
    StructField("vote", StringType(), True),
    StructField("reviewTime", StringType(), True),
    StructField("reviewerID", StringType(), True),
    StructField("asin", StringType(), True),
    StructField("reviewerName", StringType(), True),
    StructField("reviewText", StringType(), True),
    StructField("summary", StringType(), True),
    StructField("label", FloatType(), True),
])

# Read the data into a Spark DataFrame
df = spark.read.schema(json_schema_data).json(dataset_fp)

# Set the BERT model name and create a BERT tokenizer
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name, max_length=512)

# Define a Spark UDF to convert the 'label' column from float to int
def float_to_int(value):
    return int(value)


float_to_int_udf = udf(float_to_int, IntegerType())

# Apply the UDF to the DataFrame
df = df.withColumn("label", float_to_int_udf("label"))

# Define a Spark UDF to tokenize and encode the review text
def tokenize_and_encode_udf(review_text, targets):
    # Assuming max_len is defined
    max_len = 512

    # Encoded format
    encoding = tokenizer.encode_plus(
        review_text,
        add_special_tokens=True,
        max_length=max_len,
        return_token_type_ids=False,
        pad_to_max_length=True,
        return_attention_mask=True,
        return_tensors='pt',
        truncation=True
    )

    return {
        "review_text": review_text,
        "input_ids": encoding['input_ids'].flatten().tolist(),
        "attention_mask": encoding['attention_mask'].flatten().tolist(),
        "targets": targets
    }

# Define the schema for the tokenized reviews
tokenizer_schema = StructType([
    StructField("review_text", StringType(), True),
    StructField("input_ids", ArrayType(IntegerType()), True),
    StructField("attention_mask", ArrayType(IntegerType()), True),
    StructField("targets", IntegerType(), True)
])

# Apply the UDF to the DataFrame
tokenize_and_encode = udf(tokenize_and_encode_udf, tokenizer_schema)
df = df.withColumn("tokenized_reviews", tokenize_and_encode(
    col("reviewText"), col("label").cast("int")))

# Define a PyTorch module for sentiment classification
class SentimentClassifier(nn.Module):

    # Constructor class
    def __init__(self, n_classes):
        super(SentimentClassifier, self).__init__()
        self.bert = BertModel.from_pretrained(model_name)
        self.drop = nn.Dropout(p=0.5)
        self.out = nn.Linear(self.bert.config.hidden_size, n_classes)

    # Forward propagaion class
    def forward(self, input_ids, attention_mask):
        bert_output = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask
        )
        pooled_output = bert_output.pooler_output  # Accessing the pooled output
        #  Add a dropout layer
        output = self.drop(pooled_output)
        return self.out(output)

# Instantiate the SentimentClassifier model
test_model = SentimentClassifier(n_classes=3)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
test_model.to(device)

# Load a pre-trained checkpoint for the model
test_checkpoint_file = '/content/drive/MyDrive/CS5344_Project/checkpoint/checkpoint_epoch_1.pth'
test_checkpoint = torch.load(
    test_checkpoint_file, map_location=torch.device('cpu'))
test_model.load_state_dict(test_checkpoint['model_state_dict'])

# use tokenised reviews with input_ids and attention_mask to predict sentiment
# Define a Spark UDF for predicting sentiment using the tokenized reviews
def predict_sentiment(tokenized_reviews):

    with torch.no_grad():
        # Extract input_ids and attention_mask from the tokenized_reviews dictionary
        input_ids_tensor = torch.tensor(
            [tokenized_reviews["input_ids"]]).to(device)
        attention_mask_tensor = torch.tensor(
            [tokenized_reviews["attention_mask"]]).to(device)

        # Assuming your model is already loaded in the 'test_model' variable
        outputs = test_model(input_ids=input_ids_tensor,
                             attention_mask=attention_mask_tensor)
        _, preds = torch.max(outputs, dim=1)

    return preds.item()


# Define the schema for the prediction UDF output
predict_sentiment_udf = udf(predict_sentiment, IntegerType())


# Apply the UDF to the DataFrame
df = df.withColumn("prediction", predict_sentiment_udf(
    col("tokenized_reviews")))

# Select 'label' and 'prediction' columns from the DataFrame
# Coalesce(1): Write the data to a single CSV file
df.select('label', 'prediction').coalesce(1).write.mode("overwrite").csv(
    "/content/drive/MyDrive/CS5344_Project/Dataset/Prediction__bert_output.csv")
```
